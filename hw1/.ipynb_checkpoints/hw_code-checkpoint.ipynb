```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
```

```python
'''
Part A: create a blackbody function:
'''

#planck's constant
h=6.6261E-27*u.erg*u.s

#speed of light
c=3E10*u.cm/u.s

#boltzmann's constant
k=1.3807E-16*u.erg/u.K

def blackbody(wavelength,temp):
    flux_d=(2*h*c**2/wavelength**5)*1/(np.expm1(h*c/wavelength/k/temp))
    return flux_d

'''
plot the blackbody spectrum of the sun at a distance of 10 pc
'''

t_sun=5800*u.K
wave=(np.linspace(100,3000,1000)*u.nm).to_value(u.cm)*u.cm
sun_flux_d=blackbody(wave,t_sun)/u.sr
#plt.plot(wave,sun_flux_d)

d=(10*u.pc).to_value(u.cm)
r_sun=7E10*u.cm

omega_sun=4*np.pi*r_sun**2/d**2*u.sr
obs_flux=omega_sun*sun_flux_d

fig,ax=plt.subplots()
ax.set_xlabel('log Wavelength (cm)')
ax.set_ylabel('log Flux Density (erg s$^{-1}$ cm$^{-2}$ cm$^{-1}$)')

plt.plot(np.log10(wave.value),np.log10(obs_flux.value))
#plt.plot((wave.value),(obs_flux.value))
```

```python
'''
mdwarf and white dwarf companion:
'''

#plot the spectrum of the white dwarf:
r_wd=6.4E8*u.cm
t_wd=10000*u.K

wave_wd=(np.linspace(100,3000,1000)*u.nm).to_value(u.cm)*u.cm
wd_flux_d=blackbody(wave_wd,t_wd)/u.sr
omega_wd=4*np.pi*r_wd**2/d**2*u.sr
flux_wd=omega_wd*wd_flux_d
plt.plot(wave_wd, flux_wd)
plt.show()

#now plot the combined spectrum of the mdwarf and white dwarf:
plt.plot(wave_wd,flux_mdwarf+flux_wd)
plt.show()
```
